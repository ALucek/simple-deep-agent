```python
from web_search import internet_search

results = internet_search(query="best cafe in paris")

print(results)
```

```python
import tiktoken

# Choose an encoding; for OpenAI GPT-3.5/4 models, "cl100k_base" is used.
encoding = tiktoken.get_encoding("cl100k_base")

# Extract all raw_content from results
raw_contents = [result['raw_content'] for result in results['results'] if result['raw_content'] is not None]

# Count tokens for each raw_content
token_counts = []
for i, raw_content in enumerate(raw_contents):
    tokens = encoding.encode(raw_content)
    token_count = len(tokens)
    token_counts.append({
        'result_index': i,
        'url': results['results'][i]['url'],
        'token_count': token_count
    })
    print(f"Result {i} ({results['results'][i]['url'][:50]}...): {token_count:,} tokens")

# Total tokens
total_tokens = sum(item['token_count'] for item in token_counts)
print(f"\n{'='*60}")
print(f"Total tokens across all raw_content: {total_tokens:,}")
print(f"Number of results with raw_content: {len(raw_contents)}")
print(f"Average tokens per result: {total_tokens/len(raw_contents):,.0f}")
```

```python
# Count tokens for the 'content' field (shorter summaries)
content_token_counts = []
for i, result in enumerate(results['results']):
    content = result.get('content', '')
    if content:
        tokens = encoding.encode(content)
        token_count = len(tokens)
        content_token_counts.append({
            'result_index': i,
            'url': result['url'],
            'token_count': token_count
        })
        print(f"Result {i} ({result['url'][:50]}...): {token_count:,} tokens")

# Total tokens for content
total_content_tokens = sum(item['token_count'] for item in content_token_counts)
print(f"\n{'='*60}")
print(f"Total tokens in 'content' fields: {total_content_tokens:,}")
print(f"Number of results: {len(content_token_counts)}")
print(f"Average tokens per content: {total_content_tokens/len(content_token_counts):,.0f}")
```
